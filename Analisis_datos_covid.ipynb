```python
fechas = ["fecha_reporte_web","fecha_de_notificaci_n","fecha_inicio_sintomas","fecha_muerte","fecha_diagnostico","fecha_recuperado"]

for fecha in fechas:
    #print(covid[fecha])
    covid[fecha] = pd.to_datetime(covid[fecha],format="%d/%m/%Y %H:%M:%S")
```

```python
deptos = covid.groupby(["departamento_nom"])[["edad"]].describe()
deptos.columns = deptos.columns.droplevel(0)
```

```python
deptos
```

```python
f,ax = plt.subplots(figsize=(16,6))

ax.bar(
    x = deptos.index,
    height = deptos["mean"],
    color = "skyblue"
)
ax.set(title="Edad promedio x departamento")
plt.xticks(rotation=45,ha="right")
plt.show()
plt.close()
```

```python
f,ax = plt.subplots(figsize=(16,6))

ax.plot(deptos.index,deptos["mean"],'r+--') # roja

ax.set(title="Edad promedio x departamento")
plt.xticks(rotation=45,ha="right")
plt.show()
plt.close()
```
